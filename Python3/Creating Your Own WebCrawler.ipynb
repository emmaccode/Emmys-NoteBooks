```python
import requests
import re
from urllib.parse import urljoin
```

```python
url = "https://www.google.com/"
response = requests.get(url)

if response.status_code == 200:
	print(response.text) 
else:
	print(f"Error: {response.status_code}")
```

```python
import requests


class WebCrawler:
	def __init__(self, start_url, max_depth=3):
		self.start_url = start_url
		self.max_depth = max_depth
		self.visited = {}  # Stores URL as key and extracted data as value

	def fetch_page(self, url):
		"""Fetch the webpage content"""
		try:
			response = requests.get(url, timeout=5)
			if response.status_code == 200:
				return response.text
		except requests.RequestException:
			return None
		return None

	def extract_metadata(self, html):
		"""Extract title, description, and keywords using regex"""
		metadata = {
			"title": re.search(r"<title>(.*?)</title>", html, re.IGNORECASE),
			"description": re.search(r'<meta\s+name=["\']description["\']\s+content=["\'](.*?)["\']', html, re.IGNORECASE),
			"keywords": re.search(r'<meta\s+name=["\']keywords["\']\s+content=["\'](.*?)["\']', html, re.IGNORECASE)
		}
		return {key: (match.group(1) if match else None) for key, match in metadata.items()}

	def extract_links(self, html, base_url):
		"""Extract all href links from the page"""
		return [urljoin(base_url, link) for link in re.findall(r'href=["\'](https?://.*?)(?=["\'])', html, re.IGNORECASE)]

	def crawl(self, url, depth=0):
		"""Recursively crawl pages up to max_depth"""
		if depth > self.max_depth or url in self.visited:
			return
		
		print(f"Crawling: {url} (Depth {depth})")
		html = self.fetch_page(url)
		if not html:
			return
		
		self.visited[url] = self.extract_metadata(html)
		links = self.extract_links(html, url)

		# Recursively crawl each new link
		for link in links:
			self.crawl(link, depth + 1)

	def run(self):
		self.crawl(self.start_url)
		return self.visited  # Return accumulated data
```

```python
WebCrawler("https://www.google.com", max_depth=10)
crawled_data = crawler.run()
```

```python
# Print accumulated data
for url, data in crawled_data.items():
	print(f"\nURL: {url}")
	print(f"Title: {data['title']}")
	print(f"Description: {data['description']}")
	print(f"Keywords: {data['keywords']}")
```
